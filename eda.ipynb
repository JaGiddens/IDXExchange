```python
# check for duplicates
df1 = df.copy()
dupes = [col for col in df1.columns if f"{col}.1" in df1.columns] # duplicates have .1 in name
{col: df1[col].equals(df1[f"{col}.1"]) for col in dupes} # see if they are identical
```

```python
# drop duplicates
df1 = df1.drop(columns=[f"{col}.1" for col in dupes])

# check again
dupes = [col for col in df1.columns if f"{col}.1" in df1.columns]
{col: df1[col].equals(df1[f"{col}.1"]) for col in dupes} 
```

```python
print("original shape:", (df.shape))
print("modified shape:", (df1.shape))
```
